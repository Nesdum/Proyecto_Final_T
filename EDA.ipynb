```python
# Calcular el rendimiento diario
df['Daily Return'] = df['Close'].pct_change()

# Calcular el rendimiento acumulado
df['Cumulative Return'] = (1 + df['Daily Return']).cumprod()

# Calcular el rendimiento promedio
average_daily_return = df['Daily Return'].mean()
annualized_return = (1 + average_daily_return) ** 252 - 1  # Suponiendo 252 días hábiles en un año

# Mostrar las medidas de rentabilidad
print("Rendimiento promedio diario:", average_daily_return)
print("Rendimiento anualizado:", annualized_return)
print("\nPrimeros 5 datos con rentabilidad:")
print(df[['Close', 'Daily Return', 'Cumulative Return']].head())
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(df['Close'], shade=True, color='blue')
plt.xlabel('Precio de Cierre (Close)')
plt.ylabel('Densidad')
plt.title('Kernel Density Estimation (KDE) del Precio de Cierre')
plt.grid()
plt.show()
```

```python
# Calcular la desviación estándar móvil (volatilidad móvil) con una ventana de 30 días
rolling_std = df['Close'].rolling(window=30).std()

# Graficar la volatilidad móvil
plt.figure(figsize=(12, 6))
plt.plot(df.index, rolling_std, label='Volatilidad Móvil (30 días)', color='orange')
plt.xlabel('Fecha')
plt.ylabel('Desviación Estándar')
plt.title('Volatilidad Móvil (Rolling Std) - 30 días')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calcular las medias móviles simples (SMA)
sma_30 = df['Close'].rolling(window=30).mean()
sma_100 = df['Close'].rolling(window=100).mean()
sma_200 = df['Close'].rolling(window=200).mean()

# Graficar el precio de cierre y las medias móviles
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Precio de Cierre', color='blue', alpha=0.6)
plt.plot(df.index, sma_30, label='SMA (30 días)', color='orange')
plt.plot(df.index, sma_100, label='SMA (100 días)', color='green')
plt.plot(df.index, sma_200, label='SMA (200 días)', color='red')

# Configurar etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.title('Media Móvil de Precios (SMA) - 30, 100, 200 días')
plt.legend()
plt.grid()
plt.show()
```
